## 1️⃣ Matplotlib — The 5 Charts You'll Use Most

### Rule: every Matplotlib chart follows the same pattern:
```python
fig, ax = plt.subplots(figsize=(width, height))
ax.plot(x, y)           # or bar, scatter, hist, etc.
ax.set_title('Title')
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
plt.tight_layout()
plt.show()
```

````python
# Chart 1 — LINE CHART (trends over time)
weeks = [1, 2, 3, 4, 5, 6, 7, 8]
cs555_hours = [1.5, 3.0, 3.0, 1.5, 3.0, 1.5, 1.5, 1.5]
cs569_hours = [0.5, 0.5, 1.5, 3.0, 3.0, 3.0, 3.0, 1.5]
qnt_hours   = [0,   0,   0,   0,   0,   3.0, 1.5, 1.5]

fig, ax = plt.subplots(figsize=(9, 4))
ax.plot(weeks, cs555_hours, 'o-', color='#5B8DEE', linewidth=2, label='CS 555')
ax.plot(weeks, cs569_hours, 's-', color='#2EB87A', linewidth=2, label='CS 569')
ax.plot(weeks, qnt_hours,   '^-', color='#E8A020', linewidth=2, label='QNT 745')
ax.set_title('Prep Hours Per Week by Course', fontsize=13, pad=10)
ax.set_xlabel('Week')
ax.set_ylabel('Hours')
ax.legend()
ax.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()
````

````python
# Chart 2 — BAR CHART (comparisons)
severities = ['Critical', 'High', 'Medium', 'Low']
counts     = [8, 14, 22, 6]
colors     = ['#E24B4A', '#F5B731', '#5B8DEE', '#3ECF8E']

fig, ax = plt.subplots(figsize=(7, 4))
bars = ax.bar(severities, counts, color=colors, edgecolor='white', linewidth=0.8)
ax.set_title('Vulnerabilities by Severity', fontsize=13, pad=10)
ax.set_ylabel('Count')
# Add value labels on top of bars
for bar, val in zip(bars, counts):
    ax.text(bar.get_x() + bar.get_width()/2, bar.get_height() + 0.3,
            str(val), ha='center', va='bottom', fontsize=11)
plt.tight_layout()
plt.show()
````

````python
# Chart 3 — HISTOGRAM (data distribution)
np.random.seed(42)
cvss_scores = np.concatenate([
    np.random.normal(9.2, 0.3, 30),   # critical cluster
    np.random.normal(7.5, 0.5, 50),   # high cluster
    np.random.normal(5.0, 1.0, 80),   # medium cluster
])
cvss_scores = np.clip(cvss_scores, 0, 10)

fig, ax = plt.subplots(figsize=(8, 4))
ax.hist(cvss_scores, bins=20, color='#5B8DEE', edgecolor='white', alpha=0.85)
ax.axvline(9.0, color='#E24B4A', linestyle='--', linewidth=1.5, label='Critical threshold (9.0)')
ax.axvline(7.0, color='#F5B731', linestyle='--', linewidth=1.5, label='High threshold (7.0)')
ax.set_title('CVSS Score Distribution', fontsize=13, pad=10)
ax.set_xlabel('CVSS Score')
ax.set_ylabel('Number of Vulnerabilities')
ax.legend()
plt.tight_layout()
plt.show()
````

````python
# Chart 4 — SCATTER PLOT (relationships between two variables)
np.random.seed(7)
n = 60
age        = np.random.randint(20, 65, n)
cvss       = 10 - (age / 65) * 3 + np.random.randn(n) * 1.5
cvss       = np.clip(cvss, 0, 10)
patched    = np.random.choice([True, False], n)

fig, ax = plt.subplots(figsize=(7, 5))
colors_scatter = ['#3ECF8E' if p else '#E24B4A' for p in patched]
ax.scatter(age, cvss, c=colors_scatter, alpha=0.7, s=60, edgecolors='white', linewidth=0.5)
ax.set_title('System Age vs CVSS Score', fontsize=13, pad=10)
ax.set_xlabel('System Age (months)')
ax.set_ylabel('CVSS Score')
from matplotlib.patches import Patch
legend_handles = [Patch(color='#3ECF8E', label='Patched'), Patch(color='#E24B4A', label='Unpatched')]
ax.legend(handles=legend_handles)
plt.tight_layout()
plt.show()
````

````python
# Chart 5 — SUBPLOTS (multiple charts in one figure — used in ML evaluation)
fig, axes = plt.subplots(1, 3, figsize=(13, 4))
fig.suptitle('Risk Dashboard — Overview', fontsize=13)

# Plot 1: Pie chart
ax1 = axes[0]
sizes  = [8, 14, 22, 6]
labels = ['Critical', 'High', 'Medium', 'Low']
clrs   = ['#E24B4A', '#F5B731', '#5B8DEE', '#3ECF8E']
ax1.pie(sizes, labels=labels, colors=clrs, autopct='%1.0f%%', startangle=90)
ax1.set_title('Severity Mix')

# Plot 2: Horizontal bar
ax2 = axes[1]
teams   = ['Team A', 'Team B', 'Team C', 'Team D']
open_ct = [12, 5, 8, 3]
ax2.barh(teams, open_ct, color='#9B7EF5')
ax2.set_title('Open Vulns by Team')
ax2.set_xlabel('Count')

# Plot 3: Line trend
ax3 = axes[2]
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun']
score  = [62, 68, 71, 75, 74, 80]
ax3.plot(months, score, 'o-', color='#2EB87A', linewidth=2)
ax3.fill_between(range(6), score, alpha=0.15, color='#2EB87A')
ax3.set_title('Compliance Score Trend')
ax3.set_ylabel('Score')
ax3.set_xticks(range(6)); ax3.set_xticklabels(months)

plt.tight_layout()
plt.show()
print('✅ This is a GRC risk dashboard — exactly what QNT 745 will teach you to build in Tableau!')
````

## 2️⃣ Scikit-learn — The fit/predict Pattern

Scikit-learn makes ML incredibly simple with one consistent API:
```python
model = SomeAlgorithm()    # 1. Choose algorithm
model.fit(X_train, y_train) # 2. Train on data
predictions = model.predict(X_test) # 3. Predict
```

That's it. Every algorithm in CS 569 follows this exact pattern.

````python
# Load the famous Iris dataset — 150 flower samples, 4 features, 3 species
iris = load_iris()
X = iris.data    # features: sepal length, sepal width, petal length, petal width
y = iris.target  # labels: 0=setosa, 1=versicolor, 2=virginica

print('Features shape:', X.shape)   # (150, 4)
print('Labels shape:', y.shape)     # (150,)
print('Feature names:', iris.feature_names)
print('Classes:', iris.target_names)
print('First 5 rows:\n', X[:5])
````

````python
# Step 1: Split data into train and test sets
# CRITICAL RULE: NEVER train and test on the same data!
X_train, X_test, y_train, y_test = train_test_split(
    X, y,
    test_size=0.2,    # 20% goes to test, 80% to train
    random_state=42   # reproducibility
)
print(f'Train size: {X_train.shape[0]} samples')
print(f'Test size:  {X_test.shape[0]} samples')
````

````python
# Step 2: Scale features — StandardScaler makes all features mean=0, std=1
# WHY: many algorithms are sensitive to scale (distance-based models, etc.)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)   # fit AND transform on train
X_test_scaled  = scaler.transform(X_test)         # ONLY transform on test!

print('Before scaling — mean of first feature:', round(X_train[:, 0].mean(), 2))
print('After scaling  — mean of first feature:', round(X_train_scaled[:, 0].mean(), 4))
print()
print('⚠️  KEY RULE: Always fit scaler on TRAIN data, then transform TEST data.')
print('   Fitting on test data = data leakage = cheating!')
````

````python
# Step 3: Train a classifier
from sklearn.neighbors import KNeighborsClassifier

model = KNeighborsClassifier(n_neighbors=5)
model.fit(X_train_scaled, y_train)           # train

# Step 4: Predict
y_pred = model.predict(X_test_scaled)        # predict on unseen test data

# Step 5: Evaluate
accuracy = accuracy_score(y_test, y_pred)
print(f'Accuracy: {accuracy * 100:.1f}%')
print()
print('Predictions:', y_pred)
print('Actual:     ', y_test)
````

````python
# Confusion Matrix — shows what the model got right and wrong
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=iris.target_names)
fig, ax = plt.subplots(figsize=(5, 4))
disp.plot(ax=ax, colorbar=False, cmap='Blues')
ax.set_title('Confusion Matrix — KNN on Iris')
plt.tight_layout()
plt.show()
print('Rows = actual class | Cols = predicted class')
print('Diagonal = correctly classified ✅')
````
