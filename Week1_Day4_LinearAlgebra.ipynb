## 1️⃣ Vectors — The Building Blocks

A **vector** is an ordered list of numbers representing a point or direction in space.
- In physics: an arrow with direction and magnitude
- In CS/ML: a list of feature values for one data point
- Example: one Iris flower = `[5.1, 3.5, 1.4, 0.2]` = a vector in 4D space

```python
# Vectors in NumPy
v1 = np.array([2, 3])    # 2D vector
v2 = np.array([1, -1])   # another 2D vector

# Vector addition (geometrically: place tip of v2 at tail of v1)
v_sum = v1 + v2
print('v1 + v2 =', v_sum)

# Scalar multiplication (scales the vector)
print('2 * v1 =', 2 * v1)
print('0.5 * v1 =', 0.5 * v1)

# Magnitude (length) of a vector
magnitude_v1 = np.linalg.norm(v1)
print(f'|v1| = sqrt({v1[0]}^2 + {v1[1]}^2) = {magnitude_v1:.3f}')
```

```python
# Visualize 2D vectors (like 3Blue1Brown!)
fig, ax = plt.subplots(figsize=(6, 6))
origin = [0, 0]

# Draw vectors as arrows from origin
ax.quiver(*origin, *v1, color='#5B8DEE', scale=1, scale_units='xy', angles='xy', width=0.012, label='v1 [2,3]')
ax.quiver(*origin, *v2, color='#E24B4A', scale=1, scale_units='xy', angles='xy', width=0.012, label='v2 [1,-1]')
ax.quiver(*origin, *v_sum, color='#2EB87A', scale=1, scale_units='xy', angles='xy', width=0.012, label=f'v1+v2 {list(v_sum)}')

# Draw parallelogram to show addition
ax.quiver(*v1, *v2, color='#E24B4A', alpha=0.3, scale=1, scale_units='xy', angles='xy', width=0.008)
ax.quiver(*v2, *v1, color='#5B8DEE', alpha=0.3, scale=1, scale_units='xy', angles='xy', width=0.008)

ax.set_xlim(-1, 5); ax.set_ylim(-2, 5)
ax.axhline(0, color='gray', linewidth=0.5)
ax.axvline(0, color='gray', linewidth=0.5)
ax.grid(alpha=0.3)
ax.legend(loc='upper left')
ax.set_title('Vector Addition — v1 + v2 = v_sum', fontsize=12)
plt.tight_layout(); plt.show()
```

## 2️⃣ Dot Product — The Most Important ML Operation

**Dot product** = multiply element-wise, then sum everything.

`v1 · v2 = v1[0]*v2[0] + v1[1]*v2[1] + ... `

Why it matters in ML:
- Measures **similarity** between two vectors (how aligned they are)
- The core operation inside neural networks (every layer is a dot product)
- Used in SVMs, PCA, attention mechanisms in transformers

```python
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# Manual dot product
manual = 1*4 + 2*5 + 3*6
print(f'Manual: {manual}')

# NumPy dot product
numpy_dot = np.dot(a, b)
print(f'np.dot: {numpy_dot}')

# @ operator (matrix multiply — same as dot for 1D)
at_op = a @ b
print(f'a @ b:  {at_op}')

# Geometric interpretation: dot product relates to cosine of angle between vectors
cos_angle = np.dot(a, b) / (np.linalg.norm(a) * np.linalg.norm(b))
angle_deg  = np.degrees(np.arccos(cos_angle))
print(f'Angle between a and b: {angle_deg:.1f}°')
print('(0° = same direction, 90° = perpendicular, 180° = opposite)')
```

```python
# Dot product as similarity — crucial in ML!
query = np.array([0.9, 0.1, 0.5])   # e.g. a search query embedding

doc1  = np.array([0.8, 0.2, 0.4])   # similar document
doc2  = np.array([0.1, 0.9, 0.0])   # very different document
doc3  = np.array([0.9, 0.1, 0.6])   # almost identical

print('Similarity scores (dot product):')
print(f'  query vs doc1: {np.dot(query, doc1):.3f}')
print(f'  query vs doc2: {np.dot(query, doc2):.3f}  ← lowest (most different)')
print(f'  query vs doc3: {np.dot(query, doc3):.3f}  ← highest (most similar)')
print()
print('This is how search engines and recommendation systems work!')
```

## 3️⃣ Matrix Multiplication — Linear Transformations

A **matrix** transforms (rotates, scales, reflects) a vector into a new vector.

Key rule: `(m×n) @ (n×p) = (m×p)` — inner dimensions must match!

```python
A = np.array([[2, 0],
              [0, 3]])   # scaling matrix (x2 in x, x3 in y)
v = np.array([1, 1])

# Transform v using A
v_transformed = A @ v
print(f'Original vector: {v}')
print(f'After scaling matrix: {v_transformed}')

# Rotation matrix by 90 degrees
theta = np.pi / 2   # 90 degrees in radians
R = np.array([[np.cos(theta), -np.sin(theta)],
              [np.sin(theta),  np.cos(theta)]])
print(f'
Rotation matrix (90°):\n{np.round(R, 2)}')
print(f'[1,0] rotated 90°: {np.round(R @ np.array([1, 0]), 2)}  → should be [0,1]')
```

```python
# Matrix multiplication — MUST KNOW the rule
A = np.array([[1, 2, 3],
              [4, 5, 6]])    # shape: (2, 3)

B = np.array([[7,  8],
              [9,  10],
              [11, 12]])     # shape: (3, 2)

C = A @ B    # result shape: (2, 2) — inner dims (3) must match!
print('A (2x3) @ B (3x2) = C (2x2):')
print(C)

# Manual verification of C[0,0]: row 0 of A · col 0 of B
manual_00 = 1*7 + 2*9 + 3*11
print(f'
C[0,0] manual: {manual_00} | NumPy: {C[0,0]}  ✅')
```

```python
# Visualize transformation — see 3B1B Chapter 3 come alive!
fig, axes = plt.subplots(1, 2, figsize=(11, 5))

# Define a set of vectors (a grid)
def draw_grid_transformed(ax, M, title, color='#5B8DEE'):
    # Draw basis vectors
    origin = [0, 0]
    i_hat = np.array([1, 0])
    j_hat = np.array([0, 1])
    ax.quiver(*origin, *(M @ i_hat), color='#E24B4A', scale=1, scale_units='xy', angles='xy', width=0.015, label='i-hat (transformed)')
    ax.quiver(*origin, *(M @ j_hat), color='#2EB87A', scale=1, scale_units='xy', angles='xy', width=0.015, label='j-hat (transformed)')
    # Draw a few sample points transformed
    pts = np.array([[1,0],[0,1],[1,1],[2,0],[0,2],[-1,0],[0,-1]])
    t_pts = (M @ pts.T).T
    ax.scatter(t_pts[:,0], t_pts[:,1], color=color, s=60, zorder=5)
    ax.set_xlim(-3,4); ax.set_ylim(-3,4)
    ax.axhline(0,color='gray',lw=0.5); ax.axvline(0,color='gray',lw=0.5)
    ax.grid(alpha=0.3); ax.legend(fontsize=8)
    ax.set_title(title, fontsize=11)
    ax.set_aspect('equal')

# Identity (no transformation)
I = np.eye(2)
draw_grid_transformed(axes[0], I, 'Identity matrix (no change)')

# Shear transformation
M = np.array([[1, 1], [0, 1]])
draw_grid_transformed(axes[1], M, 'Shear matrix [[1,1],[0,1]]', '#9B7EF5')

plt.tight_layout(); plt.show()
```

## 4️⃣ Determinant & Inverse

**Determinant**: scalar measure of how much a matrix scales area.
- det = 0 → matrix collapses space (singular, not invertible)
- det > 0 → preserves orientation
- det < 0 → flips orientation

**Inverse**: the "undo" matrix. A⁻¹ @ A = I (identity)

```python
A = np.array([[3, 1],
              [2, 4]])

det_A = np.linalg.det(A)
print(f'det(A) = {det_A:.1f}   → matrix scales area by {det_A:.1f}x')

inv_A = np.linalg.inv(A)
print(f'\nA inverse:\n{np.round(inv_A, 4)}')

# Verify: A @ A_inv should equal identity
product = A @ inv_A
print(f'\nA @ A_inv (should be identity):\n{np.round(product, 10)}')

# Singular matrix — det = 0, no inverse
singular = np.array([[2, 4],
                     [1, 2]])
print(f'\ndet(singular) = {np.linalg.det(singular):.1f}  → cannot invert!')
```

## 5️⃣ Eigenvalues & Eigenvectors

**Key insight (from 3Blue1Brown Ch. 14):** Most vectors get knocked off their span when transformed. But **eigenvectors** stay on their own span — they only get scaled by their **eigenvalue**.

**Why this matters for GRC/ML:** PCA (Week 7) uses eigenvalues to find the directions of maximum variance in your data. You'll use this to compress vulnerability datasets while keeping the most risk-relevant information.

```python
A = np.array([[3, 1],
              [0, 2]])

eigenvalues, eigenvectors = np.linalg.eig(A)
print('Eigenvalues:', eigenvalues)
print('Eigenvectors (columns):\n', eigenvectors)

# Verify: A @ v = λ * v  for each eigenpair
for i in range(len(eigenvalues)):
    lam = eigenvalues[i]
    v   = eigenvectors[:, i]
    Av  = A @ v
    lv  = lam * v
    print(f'\nEigenpair {i}: λ={lam:.1f}, v={np.round(v,3)}')
    print(f'  A @ v  = {np.round(Av, 6)}')
    print(f'  λ * v  = {np.round(lv, 6)}')
    print(f'  Match: {np.allclose(Av, lv)}  ✅')
```

```python
# Intuitive demo: which direction has the most variance in data?
# (preview of what PCA does in Week 7)
np.random.seed(42)
# Correlated 2D data (like two related vulnerability metrics)
x = np.random.randn(100)
y = 2 * x + np.random.randn(100) * 0.5   # y correlates strongly with x

data = np.column_stack([x, y])
cov_matrix = np.cov(data.T)   # covariance matrix
eigenvals, eigenvecs = np.linalg.eig(cov_matrix)

print('Covariance matrix:\n', np.round(cov_matrix, 2))
print(f'\nEigenvalues: {np.round(eigenvals, 2)}')
print('Eigenvectors:\n', np.round(eigenvecs, 3))
print(f'\nThe first eigenvector {np.round(eigenvecs[:,0],2)} points in the direction of most variance.')
print('This is exactly what PCA finds! (Week 7)')

fig, ax = plt.subplots(figsize=(6, 5))
ax.scatter(data[:,0], data[:,1], alpha=0.4, color='#5B8DEE', s=20)
# Draw eigenvectors scaled by eigenvalues
for i in range(2):
    ev = eigenvecs[:, i] * np.sqrt(eigenvals[i]) * 2
    ax.quiver(0, 0, ev[0], ev[1], color=['#E24B4A','#2EB87A'][i],
              scale=1, scale_units='xy', angles='xy', width=0.015,
              label=f'PC{i+1} (λ={eigenvals[i]:.1f})')
ax.set_title('PCA Preview — Eigenvectors of Covariance Matrix', fontsize=11)
ax.legend(); ax.grid(alpha=0.3); ax.set_aspect('equal')
plt.tight_layout(); plt.show()
```
